```python
from sentence_transformers import SentenceTransformer

# Load the pre-trained Sentence-BERT model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Generate embeddings for course titles and descriptions
data['Title_Embeddings'] = data['CourseTitle'].apply(lambda x: model.encode(x))
data['Description_Embeddings'] = data['Description'].apply(lambda x: model.encode(x))

# Optional: Save the embeddings to a new CSV for later use
data.to_csv('courses_with_embeddings.csv', index=False)
```
